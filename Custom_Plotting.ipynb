```python
#pip install plotly==3.10.0
#pip install cufflinks
#pip install tkinter

import numpy as np
import pandas as pd
import plotly.plotly as pl
import plotly.offline as po
import cufflinks as cf
from tkinter import filedialog
import os
```

```python
po.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
import tkinter as tk
csv = int(input("Press 1 to Select CSV data file you want to plot \n"))

if csv == 1:
    application_window = tk.Tk()

    my_filetypes = [('csv files only!', '.csv')]
    answer = filedialog.askopenfilename(parent=application_window,
                                        initialdir=os.getcwd(),
                                        title="Please select a csv file:",
                                        filetypes=my_filetypes)

    application_window.destroy()
    x = pd.read_csv(answer)
    df = pd.DataFrame(x)
    
else:
    print("Invalid Input!")
```

```python
def main(plot_type):
    if plot_type==1:
        print("Please Select the type of Plot :\n")
        print("Press 1 for Line Plot\n")
        print("Press 2 for Scatter Plot\n")
        print("Press 3 for Bar Plot\n")
        print("Press 4 for Histogram\n")
        print("Press 5 for Box Plot\n")
        full_plot= int(input())
        output = plot_data(full_plot)

    elif plot_type==2:
        print("Please Select the type of Plot :\n")
        print("Press 1 for Line Plot\n")
        print("Press 2 for Scatter Plot\n")
        print("Press 3 for Bar Plot\n")
        print("Press 4 for Histogram\n")
        print("Press 5 for Box Plot\n")
        col_plot= int(input())
        output = plot_data1(col_plot)

    else:
        print("Invalid input!!")
```

```python
def plot_data(full_data):
    if full_data==1:
        final=df.iplot(kind='scatter')
    elif full_data==2:
         final=df.iplot(kind='scatter',mode='markers')
    elif full_data==3:
        final=df.iplot(kind='bar')
    elif full_data==4:
        final=df.iplot(kind='hist')
    elif full_data==5:
        final=df.iplot(kind='box')
    else:
        print("Please Select only from 1 to 5!")
    return final

    
```

```python
def plot_data1(col_plot):
    print("Enter Number of Columns you want to plot [Upto 3 only!]\n")
    col=int(input())
    if col==1:
        colm= input("Enter The Column Name CORRECTLY!\n")
        if col_plot==1:
            final=df[colm].iplot(kind='scatter')
        elif col_plot==2:
            final=df[colm].iplot(kind='scatter',mode='markers',symbol='x',colorscale='paired')
        elif col_plot==3:
            final=df[colm].iplot(kind='bar')
        elif col_plot==4:
            final=df[colm].iplot(kind='hist')
        elif col_plot==5:
            final=df[colm].iplot(kind='box')
        else:
            final = print("Please Select between 1 to 5!")
            
    elif col==2:
        x= input("Enter The FIRST Column Name CORRECTLY!\n")
        y= input("Enter The SECOND Column Name CORRECTLY!\n")
        if col_plot==1:
            final=df[[x,y]].iplot(kind='scatter')
        elif col_plot==2:
            final=df[[x,y]].iplot(kind='scatter',mode='markers',symbol='x',colorscale='paired')
        elif col_plot==3:
            final=df[[x,y]].iplot(kind='bar')
        elif col_plot==4:
            final=df[[x,y]].iplot(kind='hist')
        elif col_plot==5:
            final=df[[x,y]].iplot(kind='box')
        else:
            final = print("Please Select between 1 to 5!")
    
    elif col==3:
        x= input("Enter The FIRST Column Name CORRECTLY!\n")
        y= input("Enter The SECOND Column Name CORRECTLY!\n")
        z= input("Enter The THIRD Column Name CORRECTLY!\n")
        if col_plot==1:
            final=df[[x,y,z]].iplot(kind='scatter')
        elif col_plot==2:
            final=df[[x,y,z]].iplot(kind='scatter',mode='markers',symbol='x',colorscale='paired')
        elif col_plot==3:
            final=df[[x,y,z]].iplot(kind='bar')
        elif col_plot==4:
            final=df[[x,y,z]].iplot(kind='hist')
        elif col_plot==5:
            final=df[[x,y,z]].iplot(kind='box')
        else:
            final = print("Please Select between 1 to 5!")
            
    else:
        print("Please Select 1 , 2 , or 3 ONLY!")
    return final
```
